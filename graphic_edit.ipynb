```python
import numpy as np
import ipyvolume as ipv
```

```python
xmin=x.min() ;ymin=y.min();xmax=x.max();ymax=y.max();
ymax
```

```python
ipv.quickscatter(x, y, z, size=1, marker="sphere", extent=[[xmin,xmax],[ymin,ymax],[-3000,2500]])
```

```python
ipv.figure()
s = ipv.scatter(x, y, z, marker='sphere', size=3)
ipv.xlim(xmin,xmax)
ipv.ylim(ymin,ymax)
ipv.zlim(-4000,-1500)
ipv.animation_control(s)# shows controls for animation controls
ipv.selector_default()
ipv.show()
```
